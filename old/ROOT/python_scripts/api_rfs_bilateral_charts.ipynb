### Bilateral aggs

```python
df = pd.read_csv("/home/dilan/Tradefeedr-Documentation/modules/ROOT/examples/api_rfs_bilateral/aggregates/raw/api_rfs_bilateral_aggregates_ex1.csv", index_col=0)
```

```python
df = df[df.Symbol=="USDJPY"]
df = df.sort_values(by = "DecayPM30s", ascending=False)
```

```python
fig = px.bar(df[df.OrderStatus == "F"], y= "DecayPM30s", title = "USDJPY DecayPM30s by LP and OrderStatus")
fig.show()
```

```python
px.bar(df[df.OrderStatus == "R"], y= "DecayPM30s", title = "USDJPY DecayPM30s by LP and OrderStatus")
```

```python
fig = px.bar(df, y= "DecayPM30s", color="OrderStatus", title = "USDJPY DecayPM30s by LP and OrderStatus", barmode="group")
fig
```

```python
fig.write_image("/home/dilan/Tradefeedr-Documentation/modules/ROOT/images/api_rfs_bilateral/aggregates/api_rfs_bilateral_aggregates_ex1_fig.png")
```
